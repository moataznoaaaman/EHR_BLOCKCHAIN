```python
class clinic:
    def __init__(self, clinic_id):
        self.clinic_id = clinic_id
        self.patient_counter = 0
        self.public_key = 0
        self.private_key = 1

    def enter_patient_registery(self,name,age,weight,height,sex,blood,pulse):
        print(name.get(),age.get(),weight.get(),height.get(),sex.get(),blood.get(),pulse.get())

    def enter_regular_visit(self,patient_id,blood,pulse,visit_reason,Diagonsis,Prescription,doctor_ref,lab_tests):
        print(patient_id.get(),blood.get(),pulse.get(),visit_reason.get(),Diagonsis.get(),Prescription.get(),doctor_ref.get(),lab_tests.get())

    def enter_lab_visit(self,patient_id,blood,pulse,lab_test,lab_test_res):
        print(patient_id.get(),blood.get(),pulse.get(),lab_test.get(),lab_test_res.get())

    # registering a new patient

    def patient_registery(self):
        root.destroy()
        app = Tk()
        app.geometry("500x500")
        ws = app.winfo_screenwidth() # width of the screen
        hs = app.winfo_screenheight() # height of the scree
        x = (ws/2) - (500/2)
        y = (hs/2) - (500/2)
        app.geometry('%dx%d+%d+%d' % (500, 500, x, y))

        name_txt=StringVar()
        name_txt.set("Name")
        name_cap=Label(app, textvariable=name_txt, height=4)
        name_cap.grid(row=2,column=0)
        name_en=Entry(app,width=50)
        name_en.grid(row=2,column=2)

        age_txt=StringVar()
        age_txt.set("Age")
        age_cap=Label(app, textvariable=age_txt, height=4)
        age_cap.grid(row=3,column=0)
        age_en=Entry(app,width=50)
        age_en.grid(row=3,column=2)

        weight_txt=StringVar()
        weight_txt.set("Weigt")
        weight_cap=Label(app, textvariable=weight_txt, height=4)
        weight_cap.grid(row=4,column=0)
        weight_en=Entry(app,width=50)
        weight_en.grid(row=4,column=2)

        height_txt=StringVar()
        height_txt.set("Height")
        height_cap=Label(app, textvariable=height_txt, height=4)
        height_cap.grid(row=5,column=0)
        height_en=Entry(app,width=50)
        height_en.grid(row=5,column=2)

        sex_txt=StringVar()
        sex_txt.set("Sex")
        sex_cap=Label(app, textvariable=sex_txt, height=4)
        sex_cap.grid(row=6,column=0)
        sex_en=Entry(app,width=50)
        sex_en.grid(row=6,column=2)

        blood_txt=StringVar()
        blood_txt.set("Blood preasure")
        blood_cap=Label(app, textvariable=blood_txt, height=4)
        blood_cap.grid(row=7,column=0)
        blood_en=Entry(app,width=50)
        blood_en.grid(row=7,column=2)

        pulse_txt=StringVar()
        pulse_txt.set("Pulse")
        pulse_cap=Label(app, textvariable=pulse_txt, height=4)
        pulse_cap.grid(row=8,column=0)
        pulse_en=Entry(app,width=50)
        pulse_en.grid(row=8,column=2)

        btn = Button(app, height=1, width=10, text="enter", command= lambda: self.enter_patient_registery(name_en, age_en, weight_en, height_en, sex_en, blood_en, pulse_en))
        btn.grid(row=9,column=2)
    # regular visit for a patient
    def regular_visit(self):
        root.destroy()
        app = Tk()
        app.geometry("500x600")
        ws = app.winfo_screenwidth() # width of the screen
        hs = app.winfo_screenheight() # height of the scree
        x = (ws/2) - (500/2)
        y = (hs/2) - (600/2)
        app.geometry('%dx%d+%d+%d' % (500, 600, x, y))

        patient_txt=StringVar()
        patient_txt.set("Patient ID")
        patient_cap=Label(app, textvariable=patient_txt, height=4)
        patient_cap.grid(row=0,column=0)
        patient_en=Entry(app,width=50)
        patient_en.grid(row=0,column=2)

        blood_txt=StringVar()
        blood_txt.set("Blood preasure")
        blood_cap=Label(app, textvariable=blood_txt, height=4)
        blood_cap.grid(row=2,column=0)
        blood_en=Entry(app,width=50)
        blood_en.grid(row=2,column=2)

        pulse_txt=StringVar()
        pulse_txt.set("Pulse")
        pulse_cap=Label(app, textvariable=pulse_txt, height=4)
        pulse_cap.grid(row=3,column=0)
        pulse_en=Entry(app,width=50)
        pulse_en.grid(row=3,column=2)

        visit_reason_txt=StringVar()
        visit_reason_txt.set("Visit reason")
        visit_reason_cap=Label(app, textvariable=visit_reason_txt, height=4)
        visit_reason_cap.grid(row=4,column=0)
        visit_reason_en=Entry(app,width=50)
        visit_reason_en.grid(row=4,column=2)

        diagnosis_txt=StringVar()
        diagnosis_txt.set("Diagonsis")
        Diagonsis_cap=Label(app, textvariable=diagnosis_txt, height=4)
        Diagonsis_cap.grid(row=5,column=0)
        Diagonsis_en=Entry(app,width=50)
        Diagonsis_en.grid(row=5,column=2)

        Prescription_txt=StringVar()
        Prescription_txt.set("Prescription (if any)")
        Prescription_cap=Label(app, textvariable=Prescription_txt, height=4)
        Prescription_cap.grid(row=6,column=0)
        Prescription_en=Entry(app,width=50)
        Prescription_en.grid(row=6,column=2)

        doctor_ref_txt=StringVar()
        doctor_ref_txt.set("Referrals to doctors (if any)")
        doctor_ref_cap=Label(app, textvariable=doctor_ref_txt, height=4)
        doctor_ref_cap.grid(row=7,column=0)
        doctor_ref_en=Entry(app,width=50)
        doctor_ref_en.grid(row=7,column=2)

        lab_test_txt=StringVar()
        lab_test_txt.set("Lab tests (if any)")
        lab_test_cap=Label(app, textvariable=lab_test_txt, height=4)
        lab_test_cap.grid(row=8,column=0)
        lab_test_en=Entry(app,width=50)
        lab_test_en.grid(row=8,column=2)
        
        btn = Button(app, height=1, width=10, text="enter", command= lambda: self.enter_regular_visit(patient_en, blood_en, pulse_en, visit_reason_en, Diagonsis_en, Prescription_en, doctor_ref_en, lab_test_en))                                                                                               
        btn.grid(row=9,column=2)

    def lab_visit(self):
        root.destroy()
        app = Tk()
        app.geometry("500x400")
        ws = app.winfo_screenwidth() # width of the screen
        hs = app.winfo_screenheight() # height of the scree
        x = (ws/2) - (500/2)
        y = (hs/2) - (400/2)
        app.geometry('%dx%d+%d+%d' % (500, 400, x, y))  

        patient_txt=StringVar()
        patient_txt.set("Patient ID")
        patient_cap=Label(app, textvariable=patient_txt, height=4)
        patient_cap.grid(row=0,column=0)
        patient_en=Entry(app,width=50)
        patient_en.grid(row=0,column=2)
        

        blood_txt=StringVar()
        blood_txt.set("Blood preasure")
        blood_cap=Label(app, textvariable=blood_txt, height=4)
        blood_cap.grid(row=2,column=0)
        blood_en=Entry(app,width=50)
        blood_en.grid(row=2,column=2)

        pulse_txt=StringVar()
        pulse_txt.set("Pulse")
        pulse_cap=Label(app, textvariable=pulse_txt, height=4)
        pulse_cap.grid(row=3,column=0)
        pulse_en=Entry(app,width=50)
        pulse_en.grid(row=3,column=2)

        lab_test_txt=StringVar()
        lab_test_txt.set("Lab test type")
        lab_test_cap=Label(app, textvariable=lab_test_txt, height=4)
        lab_test_cap.grid(row=4,column=0)
        lab_test_en=Entry(app,width=50)
        lab_test_en.grid(row=4,column=2)

        lab_test_res_txt=StringVar()
        lab_test_res_txt.set("Lab test result")
        lab_test_res_cap=Label(app, textvariable=lab_test_res_txt, height=4)
        lab_test_res_cap.grid(row=5,column=0)
        lab_test_res_en=Entry(app,width=50)
        lab_test_res_en.grid(row=5,column=2)

        btn = Button(app, height=1, width=10, text="enter", command= lambda: self.enter_lab_visit(patient_en, blood_en, pulse_en, lab_test_en, lab_test_res_en))                                                                                               
        btn.grid(row=9,column=2)
```
